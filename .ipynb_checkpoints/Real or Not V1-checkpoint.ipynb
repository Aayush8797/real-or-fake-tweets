```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.ensemble import RandomForestClassifier,GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix
import warnings
warnings.simplefilter("ignore")
from nltk.tokenize import word_tokenize
import nltk
import re
import string
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))
from nltk.stem import WordNetLemmatizer
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from bs4 import BeautifulSoup
import requests
```

#### Reading Data 

```python
train = pd.read_csv(r"C:\Users\aayus\Desktop\Kaggle\Real or Fake\train.csv")
test = pd.read_csv(r"C:\Users\aayus\Desktop\Kaggle\Real or Fake\test.csv")
submission = pd.read_csv(r"C:\Users\aayus\Desktop\Kaggle\Real or Fake\sample_submission.csv")
train = train.fillna('')
train.head(1)
```

```python
stop_words = stop_words.union(set(['http','https','com','co','this','like','amp','via']))
```

```python
def cleaning_text(text):
    lemmatizer = WordNetLemmatizer()
    lem = lambda x : lemmatizer.lemmatize(x)
    text = text.lower().strip()
    text = re.sub('\[.*?\]',' ',text)
    text = re.sub('(%s)' % re.escape(string.punctuation),'',text)
    text = re.sub('\w*\d\w*',' ',text)
    text = re.sub('#',' ',text)
    text = re.sub('\'',' ',text)
    text = re.sub('@',' ',text)
    text = re.sub(':',' ',text)
    text = re.sub(',',' ',text)
    text = re.sub('\?',' ',text)
    text = re.sub('\_',' ',text)
    text = re.sub('\.',' ',text)
    ps = PorterStemmer()
    stem = lambda x : ps.stem(x)
    text = list(map(stem,text.split()))
    text = map(lem,text)
    text = [i for i in text if i not in stop_words]     
    text = " ".join(text)
    text = re.sub('thi',' ',text)
    text = re.sub('co',' ',text)
    return text
```

```python
cleaning_text(train.iloc[0,3])
```

```python
train['textToClean'] = train[['keyword', 'location','text']].apply(lambda x: ' '.join(x), axis=1)
```

```python
train['cleanText'] = train.textToClean.apply(lambda x : cleaning_text(x))
```

```python
cv = CountVectorizer()
dataCV = cv.fit_transform(train.cleanText)
```

```python
data = pd.DataFrame(dataCV.toarray(),columns=cv.get_feature_names())
```

```python
Y = train.target
Y.name = 'Y'
```

```python
frame = pd.concat([data,pd.DataFrame(Y)],axis=1)
```

### Some EDA

```python
event = data.loc[frame.Y == 1,:]
nonEvent = data.loc[frame.Y==0,:]
```

```python
eventFrame = pd.DataFrame(event.apply(lambda x : x.sum(),axis=0))
Corpus = {}
for i in range(eventFrame.shape[0]):
    Corpus[eventFrame.iloc[i].name] = eventFrame.iloc[i].values[0]
```

```python
wc = WordCloud(background_color='white',colormap='inferno',max_font_size=150,random_state=45)
```

```python
plt.figure(figsize=[15,15])
wc.generate_from_frequencies(frequencies=Corpus)
plt.imshow(wc,interpolation='bilinear')
plt.title("Event")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=[20,5])
eventMostCommon = Counter(Corpus).most_common()[0:50]
names = list(dict(eventMostCommon).keys())
values = list(dict(eventMostCommon).values())
pos_y = np.array(values)
plt.bar(names,pos_y,color='darkgreen')
plt.title('Top 50 Event words')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
nonEventFrame = pd.DataFrame(nonEvent.apply(lambda x : x.sum(),axis=0))
CorpusNonEvent = {}
for i in range(nonEventFrame.shape[0]):
    CorpusNonEvent[nonEventFrame.iloc[i].name] = nonEventFrame.iloc[i].values[0]
```

```python
plt.figure(figsize=[15,15])
wc.generate_from_frequencies(frequencies=CorpusNonEvent)
plt.imshow(wc,interpolation='bilinear')
plt.title("Non Events")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=[20,5])
eventMostCommon = Counter(CorpusNonEvent).most_common()[0:50]
names = list(dict(eventMostCommon).keys())
values = list(dict(eventMostCommon).values())
pos_y = np.array(values)
plt.title('Top 50 Non Event words')
plt.bar(names,pos_y,color='darkblue')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```
